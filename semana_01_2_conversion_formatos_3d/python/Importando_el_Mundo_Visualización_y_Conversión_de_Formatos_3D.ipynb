```python
import trimesh
import open3d

import numpy as np
```

###**stl**

```python
mudkip_stl = trimesh.load("mudkip_cgtrader.stl")
```

```python
print("Numero de vertices mudkip_stl:", len(mudkip_stl.vertices))
print("Numero de Normales por vertices mudkip_stl:", len(mudkip_stl.vertex_normals))

print("Numero de Faces mudkip_stl:", len(mudkip_stl.faces))
print("Numero de Normales por cara mudkip_stl:", len(mudkip_stl.face_normals))
```

```python
mudkip_unique_vertices = np.unique(mudkip_stl.vertices, axis=0)

print("Vertices únicos:", len(mudkip_unique_vertices))
print("Vertices duplicados:", len(mudkip_stl.vertices) - len(mudkip_unique_vertices))
```

```python
#Muestra el modelo

mudkip_stl.show()
```

### **obj**

```python
squirtel_obj_scene = trimesh.load("squirtel_cdtrader.obj")
```

```python
#Si es una escena, convierte la escena a una sola malla

if isinstance(squirtel_obj_scene, trimesh.Scene):
    squirtel_obj = trimesh.util.concatenate(squirtel_obj_scene.dump())
else:
    squirtel_obj = squirtel_obj_scene
```

```python
squirtel_unique_vertices = np.unique(squirtel_obj.vertices, axis=0)

print("Vertices únicos:", len(squirtel_unique_vertices))
print("Vertices duplicados:", len(squirtel_obj.vertices) - len(squirtel_unique_vertices))
```

```python
#Muestra el modelo

squirtel_obj_scene.show()
```

### **Conversion de formatos**

#### **STL a OBJ y STL a GLB**

```python
mudkip_stl_exc = trimesh.load("mudkip_cgtrader.stl")

#Se exporta a otro formato
mudkip_stl_exc.export("Conversion_formatos/mudkip_stl_exc.obj")
mudkip_stl_exc.export("Conversion_formatos/mudkip_stl_exc.glb")
```

```python
prueba_stl_obj = trimesh.load("Conversion_formatos/mudkip_stl_exc.obj")
prueba_stl_obj.show()
```

```python
prueba_stl_glb = trimesh.load("Conversion_formatos/mudkip_stl_exc.glb")
prueba_stl_glb.show()
```

#### **OBJ STL a  y OBJ a GLB**

```python
squirtel_obj_scene = trimesh.load("squirtel_cdtrader.obj")

if isinstance(squirtel_obj_scene, trimesh.Scene):
    squirtel_obj = trimesh.util.concatenate(squirtel_obj_scene.dump())
else:
    squirtel_obj = squirtel_obj_scene

squirtel_obj.export("Conversion_formatos/squirtel_obj_exc.stl")
squirtel_obj.export("Conversion_formatos/squirtel_obj_exc.glb")
```

```python
prueba_obj_stl = trimesh.load("Conversion_formatos/squirtel_obj_exc.stl")
prueba_obj_stl.show()
```

```python
prueba_obj_glb = trimesh.load("Conversion_formatos/squirtel_obj_exc.glb")
prueba_obj_glb.show()
```

#### **GLB STL a  y GLB a OBJ**

```python
diglett_glb_scene = trimesh.load("diglett_cgtrader.glb")

if isinstance(diglett_glb_scene, trimesh.Scene):
    diglett_glb = trimesh.util.concatenate(diglett_glb_scene.dump())
else:
    diglett_glb = diglett_glb_scene

diglett_glb.export("Conversion_formatos/diglett_glb_exc.stl")
diglett_glb.export("Conversion_formatos/diglett_glb_exc.obj")
```

```python
prueba_glb_stl = trimesh.load("Conversion_formatos/diglett_glb_exc.stl")
prueba_glb_stl.show()
```

```python
prueba_glb_obj = trimesh.load("Conversion_formatos/diglett_glb_exc.obj")
prueba_glb_obj.show()
```
